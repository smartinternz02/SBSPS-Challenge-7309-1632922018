```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv(r'C:\Users\PRANEETHA\OneDrive\Desktop\Types_of_toilets1.csv')
headers=[]
regular=[]
os=[]
w=0.5


f1={"family":"cambria","color":"red","size":100}
f2={"family":"cambria","color":"black","size":100}



plt.figure(figsize=(100,30))
for i in data["ULB Name"]:
    headers.append(i[:5])
for i in data["Regular Workers"]:
    regular.append(i)
for i in data["Out Sourced Workers"]:
    os.append(i)

t_bar=np.arange(len(headers))

o_bar=[i+w for i in t_bar]



plt.bar(t_bar,regular,width=w,label="Regular Workers")
plt.bar(o_bar,os,width=w,label="Autos")



plt.xticks(t_bar+w+w/2,headers)
plt.ylim(0,500)
plt.xlabel("D I S T R I C T S",fontdict=f2)
plt.ylabel("Workers",fontdict=f2)
plt.title("Types of workers",fontdict=f1)
plt.legend()
plt.show() 
```
